```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')

print(df.to_string())
g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
df=sb.load_dataset('tips')
g=sb.FacetGrid(df,col="time",row="sex",hue="smoker")
#g=sb.FacetGrid(df,col="time",row="sex",hue="time")
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
#statistical estimation
df=sb.load_dataset('tips')
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show()
```

```python
#Linear relationship
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
#importing packages
sb.pairplot(df)
plt.show()
```

```python
df=sb.load_dataset('tips')
selected_vars=df[["total_bill","tip"]]
sb.pairplot(df,vars=selected_vars)
plt.show()
```

```python
sb.pairplot(df,hue="size")
plt.show
```

```python
#modify color palette
import seaborn as sb
df=sb.load_dataset('tips')
sb.pairplot(df,hue="size",palette="husl")
plt.show()
```

```python
#diagonal kind of plots
df=sb.load_dataset('tips')
sb.pairplot(df,diag_kind="kde")#kernal distributer estimator
plt.show()
```

```python
sb.pairplot(df,kind="reg")
plt.show()
```

```python
#controlling the markers
sb.pairplot(df,hue='sex',markers=["s","*"])
plt.show()
```

```python
#advanced customization with seabotm pairplot
g=sb.pairplot(df,hue='day')
g.fig.suptitle("pairplot of tips Dataset",y=1.02)
g.set(xticks=[],yticks=[])
plt.show()
```
